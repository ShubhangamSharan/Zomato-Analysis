```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
```

```python
print(dataframe)
```

```python
def HandleRate(value):
        value = str(value).split('/')
        value = value[0];
        return float(value)
dataframe['rate'] = dataframe['rate'].apply(HandleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
#type of restrurant
```

```python
sns.countplot(x=dataframe['listed_in(type)'], color = 'red')
plt.xlabel("type of restrurant")
plt.show()
```

```python
#Majority restraurant fall in dining category
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : grouped_data})
plt.plot(result , c = "red" , marker = "o")
plt.xlabel("type of restaurant")
plt.ylabel("votes")
plt.show()
```

```python
#Dining restaurants has recieved maximum votes
```

```python
plt.hist(dataframe['rate'],bins=5,color ="red")
plt.title("rating distribution")
plt.show()
```

```python
#Majority resturants recieve ratings from 3.5 to 4
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data, color ="red")
plt.show()
```

```python
# majority couples order 300 rupees order
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order',y = 'rate' , data = dataframe,color = "red")
plt.show()
```

```python
#offline order recieve lower rating as compared to online
```
